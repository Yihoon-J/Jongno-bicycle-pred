```python
 import pandas as pd
from sklearn.impute import KNNImputer
```

# 일자별 시간대별 대여내역 및 미세먼지 집계

```python
rentdust_2020=pd.DataFrame(columns=['대여일자','측정시간','count','SO2','CO','O3','NO2','PM10','PM25'])
rentdust_2021=pd.DataFrame(columns=['대여일자','측정시간','count','SO2','CO','O3','NO2','PM10','PM25'])
rentdust_2022=pd.DataFrame(columns=['대여일자','측정시간','count','SO2','CO','O3','NO2','PM10','PM25'])
```

```python
#2020
for i in range(12):
  lst=['01','02','03','04','05','06','07','08','09','10','11','12']
  rent_record=pd.read_csv('/content/drive/MyDrive/종로 따릉이 대여예측/dataset/b_2020'+str(lst[i])+'.csv')
  rent_record1=rent_record[['대여일자','대여소번호','측정시간']]
  rent_record2=rent_record[['대여일자','SO2','CO','O3','NO2','PM10','PM25','측정시간']]
  hourly_count=rent_record1.groupby(['대여일자','측정시간']).agg('count')
  hourly_dust=rent_record2.groupby(['대여일자','측정시간']).agg('max')
  rentdust=hourly_count.join(hourly_dust)
  rentdust.columns=['count','SO2','CO','O3','NO2','PM10','PM25']
  rentdust.reset_index(inplace=True)
  rentdust_2020=rentdust_2020.append(rentdust)
rentdust_2020['D-H']=rentdust_2020['대여일자']+'-'+rentdust_2020['측정시간'].astype(str)
rentdust_2020=rentdust_2020[['대여일자','측정시간','D-H','SO2','CO','O3','NO2','PM10','PM25','count']]
imputer0=KNNImputer(n_neighbors=2)
imputing0=rentdust_2020[['SO2','CO','O3','NO2','PM10','PM25']]
imputed0=pd.DataFrame(imputer0.fit_transform(imputing0))
imputed0.columns=[['SO2','CO','O3','NO2','PM10','PM25']]
rentdust_2020['SO2']=imputed0['SO2']
rentdust_2020['CO']=imputed0['CO']
rentdust_2020['NO2']=imputed0['NO2']
rentdust_2020['PM10']=imputed0['PM10']
rentdust_2020['PM25']=imputed0['PM25']
rentdust_2020
```

# 기상 관측치 전처리

```python
#2020
weather=pd.read_csv('/content/drive/MyDrive/종로 따릉이 대여예측/dataset/weather_2020.csv',encoding='CP949')
weather['시간']=pd.to_numeric(weather['일시'].str[11:13])
weather['날짜']=weather['일시'].str[:10]
weather.columns=['지점','지점명','일시','기온','강수량','풍속','습도','현지기압','전운량','시정','시간','날짜']
weather=weather[['날짜','시간','기온','강수량','풍속','습도','현지기압','전운량','시정']]

#강수량 결측치 0으로 대체
weather['강수량']=weather['강수량'].fillna(0)

#강수량 제외한 값 보간
imputer1=KNNImputer(n_neighbors=2)
imputing1=weather[['기온','습도','현지기압','전운량','시정']]
imputed1=pd.DataFrame(imputer1.fit_transform(imputing1))
imputed1.columns=[['기온','습도','기압','전운량','시정']]
weather['기온']=imputed1['기온']
weather['습도']=imputed1['습도']
weather['기압']=imputed1['기압']
weather['전운량']=imputed1['전운량']
weather['시정']=imputed1['시정']
weather['D-H']=weather['날짜']+'-'+weather['시간'].astype(str)
weather_2020=weather[['날짜','시간','D-H','기온','강수량','풍속','습도','현지기압','전운량','시정']]
```
